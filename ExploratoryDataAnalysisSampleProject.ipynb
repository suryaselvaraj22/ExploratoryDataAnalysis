### Exploratory Data Analysis Sample Project
Sample Dataset used: 
* CardioGoodFitness-1.csv - A CSV file containing information about the cardio workout of the members of a gym

```python
# Import relevant packages

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv("Datasets/CardioGoodFitness-1.csv")
```

```python
# Check the dataset
data.head()
```

```python
# Analytical summary of the dataset
data.describe(include='all')
```

```python
# Datatype and other characteristics of the data
data.info()
```

#### Summary of distribution for relevant variables

```python
# Histogram for the relevant variables 
data.hist(figsize=(20,30))
```

#### Relationship between categorical and continuous variable

```python
# Age distribution of gym members with respect to gender
sns.boxplot(x="Gender", y="Age", data = data) 
```

```python
# Product distribution of gym members with respect to gender
sns.boxplot(x="Product", y="Age", data = data)
```

```python
# Cross Tabulation: Product vs Gender
pd.crosstab(data['Product'], data['Gender'])
```

```python
# Visualize the count of Men and Women for each Product
sns.countplot(x="Product", hue = "Gender", data = data)
```

```python
# Factoring in Marital Status (Single vs Married) using Pivot Table
pd.pivot_table(data, index=['Product', 'Gender'], columns = ['MaritalStatus'], aggfunc=len)
```

```python
# Average of miles for combination of given variables using Pivot Table
pd.pivot_table(data, 'Miles',index=['Product', 'Gender'], columns = ['MaritalStatus'])
```

```python
# Plot the entire dataset
sns.pairplot(data)
```

```python
# Distribution based on Age
sns.distplot(data['Age'])
```

```python
# Distribution based on Income and Gender
data.hist(by='Gender', column = 'Income')
```

```python
# Distribution based on Product and Gender
data.hist(by='Gender', column = 'Product')
```

```python
# Correlation of data 
corr= data.corr()
sns.heatmap(corr, annot=True)
```
